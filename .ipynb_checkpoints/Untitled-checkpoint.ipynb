```python
import pandas as pd
hs = pd.read_csv("Datasets/bikesharing/day.csv")
hs.head()
```

```python
hs.drop(hs.columns[[0,1]],axis=1,inplace=True)
hs.info()
```

```python
hs.describe()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
hs.hist(bins=50, figsize=(20,15))
plt.savefig("attribute_histogram_plots")
plt.show()
```

```python
corr_matrix = hs.corr()
corr_matrix["cnt"].sort_values(ascending=False)
```

```python
from pandas.tools.plotting import scatter_matrix

attr = ["cnt","registered","casual","atemp"]
scatter_matrix(hs[attr],figsize=(12, 8))
plt.savefig('corr.png')
```

```python
X = hs[["casual","atemp","yr","season","mnth","weekday","workingday","holiday","hum","windspeed","weathersit"]]
Y = hs[["cnt"]]

```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
print('Liner Regression R squared: %.4f' % reg.score(X_test, y_test))
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
lin_mse = mean_squared_error(y_pred, y_test)
lin_rmse = np.sqrt(lin_mse)
print('Liner Regression RMSE: %.4f' % lin_rmse)
```
